```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression
%matplotlib inline
```

```python
dataset_url = 'https://raw.githubusercontent.com/iwanami-datascience/vol3/master/kato%26hoshino/q_data_x.csv'  
df = pd.read_csv(dataset_url)  
print(df.shape)
```

```python
df.describe()
```

```python
df.columns
```

```python
df_y = df['gamesecond']
df_z = df['cm_dummy']
df_X = df.drop(['gamesecond', 'cm_dummy', 'gamecount'], axis=1)
```

```python
df_X.columns
```

```python
X = df_X
z = df_z
y = df_y

reg = LogisticRegression().fit(X, z)

print("beta: ", reg.coef_)
print("alpha: ", reg.intercept_)
```

```python
reg.score(X, z)
```

```python
z_pred = reg.predict_proba(X)
print(z_pred[:10])
print(z[:10])
```

```python
df2 = pd.DataFrame({'z': z, 'z_pred0': z_pred[:, 0], 'z_pred1': z_pred[:, 1]})

df2.head(20)
```

```python
PS = z_pred[:, 1]

IPW1 = sum(z * y / PS) / sum(z / PS)
IPW0 = sum(((1 - z) * y) / (1 - PS)) / sum((1 - z) / (1 - PS))
ATE = IPW1 - IPW0
print('ATE: ', ATE)
```
